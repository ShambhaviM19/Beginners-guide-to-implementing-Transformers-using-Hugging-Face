```python
from transformers import pipeline
classifier= pipeline("sentiment-analysis")
```

```python
classifier("I am happy to try out the transformers library")
```

```python
classifier("I dont have many friends but I have the best friend in the world ")
```

```python
classifier("I dont like playing football")
```

```python
results=classifier(["I am lucky to have such a beautiful family","I am afraid of checking my results"])
for result in results:
  print(f"label: {result['label']} with score: {round(result['score'],4)}")
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name="lxyuan/distilbert-base-multilingual-cased-sentiments-student"
model=TFAutoModelForSequenceClassification.from_pretrained(model_name,from_pt=True)
tokenizer=AutoTokenizer.from_pretrained(model_name)
classifier=pipeline('sentiment-analysis',model=model_name,tokenizer=tokenizer)
```

```python
classifier("Kitna sunder din hai aaj")
```
